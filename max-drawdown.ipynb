```python
import os
import pandas as pd
```

```python
def cal_drawdown(ticker):
    print('Start calculating: ' + ticker)
    df = pd.read_csv('data/'+ticker+'.csv')

    max_drawdown = 0
    local_max_bookkeeping = 0
    local_min_bookkeeping = 0
    period_bookkeeping = 0
    local_max_idx_bookkeeping = 0
    
    local_max = 0
    local_max_idx = 0
    drawdowning = False
    local_min = 0
    for index, row in df.iterrows():
        ac = row['Adj Close']
        if ac > local_max and not drawdowning:
            local_max = ac
            local_max_idx = index
            continue
        if ac < local_max and not drawdowning:
            drawdowning = True
            local_min = ac
            continue
        if ac < local_min and drawdowning:
            local_min = ac
            continue
        if ac >= local_max and drawdowning:
            drawdowning = False
            drawdown = (local_max-local_min)/local_max
            
            if drawdown > max_drawdown:
                max_drawdown = drawdown
                local_max_bookkeeping = local_max
                local_min_bookkeeping = local_min
                period_bookkeeping = index-local_max_idx
                local_max_idx_bookkeeping = local_max_idx
            continue
    print('max from {0} to {1}'.format(local_max_bookkeeping, local_min_bookkeeping))
    print('max dawndown is {0}, last period is {1}'.format(max_drawdown, period_bookkeeping))
    print()
    print(df.iloc[[local_max_idx_bookkeeping]])
    print('==============')
    print()
        
```

```python
stickers = [
    '0700.HK',
    '1299.HK',
    '1398.HK',
    '2318.HK',
    '3988.HK',
    '9618.HK',
    '9988.HK',
    '510310.SS',
    'AAPL',
    'AJBU.SI',
    'AMZN',
    'ARKF',
    'ARKG',
    'ARKK',
    'ARKQ',
    'ARKW',
    'BABA',
    'C2PU.SI',
    'C38U.SI',
    'CQQQ',
    'D05.SI',
    'EDU',
    'GOOG',
    'J69U.SI',
    'M44U.SI',
    'ME8U.SI',
    'META',
    'MSFT',
    'NVDA',
    'PRNT',
    'PYPL',
    'QQQ',
    'SOXX',
    'SPY',
    'TSLA',
    'U',
    'XLV'
]

for s in stickers:
    cal_drawdown(s)
```
